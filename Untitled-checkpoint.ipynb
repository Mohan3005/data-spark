```python
import mysql.connector

conn = mysql.connector.connect(
    host='localhost',     
    user='root',
    password='1234'
)

cursor = conn.cursor()

create_database_query = "CREATE DATABASE your_database;"
cursor.execute(create_database_query)

use_database_query = "USE your_database;"
cursor.execute(use_database_query)

create_sales_table = """
CREATE TABLE sales (
    sale_id INT PRIMARY KEY,
    product_id INT,
    customer_id INT,
    sale_date DATE,
    amount DECIMAL(10, 2)
);
"""

create_products_table = """
CREATE TABLE products (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(100),
    price DECIMAL(10, 2)
);
"""

create_customers_table = """
CREATE TABLE customers (
    customer_id INT PRIMARY KEY,
    customer_name VARCHAR(100),
    email VARCHAR(100)
);
"""

cursor.execute(create_sales_table)
cursor.execute(create_products_table)
cursor.execute(create_customers_table)

conn.commit()

cursor.close()
conn.close()

print("Database and tables created successfully.")
```

```python
import mysql.connector
import pandas as pd

data = {
    'sale_id': [1, 2, 3, 4],
    'product_id': [101, 102, 103, 104],
    'customer_id': [1001, 1002, 1003, 1004],
    'sale_date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01'],
    'amount': [100.00, 150.00, 200.00 , 250.00]
}

merged_df = pd.DataFrame(data)

conn = mysql.connector.connect(
    host='localhost',     
    user='root',
    password='1234',
    database='your_database' 
)

cursor = conn.cursor()

for index, row in merged_df.iterrows():
    cursor.execute(
        """
        INSERT INTO sales (sale_id, product_id, customer_id, sale_date, amount) 
        VALUES (%s, %s, %s, %s, %s)
        """,
        (row['sale_id'], row['product_id'], row['customer_id'], row['sale_date'], row['amount'])
    )

conn.commit()

cursor.close()
conn.close()

print("Data inserted successfully.")
```
